```python
# Average Temperature Calculator

print("Welcome to Average Temperature Calculator")

# Step 1: Define a list of temperatures
temperatures = [23.5, 25.0, 22.8, 24.1, 26.3]

# Step 2: Calculate total temperature
total = sum(temperatures)

# Step 3: Count how many temperatures
count = len(temperatures)

# Step 4: Calculate average
average = total / count

# Step 5: Display result
print("Temperatures:", temperatures)
print("Total:", total)
print("Number of entries:", count)
print("Average Temperature:", round(average, 2), "¬∞C")
```

<h1 style="color: orange; font-size: 25px;">CLIENT PROJECT : Average Temperature Calculator</h1>

```python
#Average Temperature Calculator

def calculate_average_temperature():
    temperatures = []
    print("Enter temperatures (in Celsius). Type 'done' when finished:")

    # Take user input
    entry = input("Enter temperature: ")
    while entry.lower() != 'done':
        if entry.replace('.', '', 1).isdigit():
            temperatures.append(float(entry))
        else:
            print("Please enter a valid number.")
        entry = input("Enter temperature: ")

    # Check if list is not empty
    if len(temperatures) > 0:
        total = sum(temperatures)
        average = total / len(temperatures)
        highest = max(temperatures)
        lowest = min(temperatures)

        # Show results
        print("\n--- Temperature Report ---")
        print("Temperatures:", temperatures)
        print("Total:", round(total, 2), "¬∞C")
        print("Count:", len(temperatures))
        print("Average:", round(average, 2), "¬∞C")
        print("Highest:", highest, "¬∞C")
        print("Lowest:", lowest, "¬∞C")
        print("----------------------------")
    else:
        print("No temperatures entered.")

# Run the program
print("üå°Ô∏è Welcome to the Temperature Analyzer üå°Ô∏è")
calculate_average_temperature()
```

<h1 style="color: orange; font-size: 25px;">1. Sum of squares using list and function</h1>

```python
def sum_of_squares(nums):
    return sum([x**2 for x in nums])

numbers = [1, 2, 3, 4]
print("Sum of squares:", sum_of_squares(numbers))  # 30
```

<h1 style="color: orange; font-size: 25px;">2. Filter even numbers using lambda and filter()</h1>

```python
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print("Even numbers:", even_numbers)  # [2, 4, 6]
```

<h1 style="color: orange; font-size: 30px;">CLIENT PROJECT: Data Cleaning Script
Script to remove duplicates and filter numbers > 10</h1>


```python
# Data Cleaning Script

def clean_data(data):
    unique_data = list(set(data))  # remove duplicates
    filtered_data = [x for x in unique_data if x > 10]  # keep only numbers > 10
    return filtered_data

# Example data
raw_data = [12, 5, 12, 18, 7, 18, 3, 25]

# Clean the data
cleaned = clean_data(raw_data)
print("Original:", raw_data)
print("Cleaned:", cleaned)
```

<h1 style="color: orange; font-size: 25px;">2. Pandas: DataFrame Creation & Basic Operations</h1>

```python
import pandas as pd

# Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Math': [85, 90, None, 70],
    'Science': [78, 88, 92, None]
}

df = pd.DataFrame(data)

# View basic info
print("Original DataFrame:")
print(df)

# Drop rows with missing values
cleaned_df = df.dropna()

# Calculate average scores (row-wise)
cleaned_df['Average'] = cleaned_df[['Math', 'Science']].mean(axis=1)

print("\nCleaned DataFrame with Averages:")
print(cleaned_df)
```

<h1 style="color: orange; font-size: 30px;">Client Project: Clean and Aggregate Dataset</h1>

```python
import pandas as pd

# Sample dataset
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Class': ['A', 'B', 'A', 'B', 'A'],
    'Math': [85, 90, None, 70, 95],
    'Science': [78, 88, 92, None, 80]
}

df = pd.DataFrame(data)

# Step 1: Fill missing values with column mean
df['Math'].fillna(df['Math'].mean(), inplace=True)
df['Science'].fillna(df['Science'].mean(), inplace=True)

# Step 2: Add average score column
df['Average'] = df[['Math', 'Science']].mean(axis=1)

# Step 3: Group by class and calculate class-wise average
grouped = df.groupby('Class')[['Math', 'Science', 'Average']].mean()

# Print results
print("Cleaned and Aggregated Data:")
print(grouped)
```

<h1 style="color: orange; font-size: 30px;">Hands-On: Visualizations for Dataset Analysis</h1>

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
iris = sns.load_dataset("iris")

# 1. Line Plot (Sepal Length of first 30 flowers)
plt.plot(iris['sepal_length'][:30])
plt.title("Sepal Length of First 30 Flowers")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.grid(True)
plt.show()

# 2. Histogram (Petal Length distribution)
plt.hist(iris['petal_length'], bins=15, color='lightgreen', edgecolor='black')
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()

# 3. Boxplot (Comparison between species)
sns.boxplot(x="species", y="sepal_width", data=iris)
plt.title("Sepal Width Comparison by Species")
plt.show()
```

<h1 style="color: orange; font-size: 30px;">Client Project: Dashboard ‚Äì Visualizing Relationships</h1>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
iris = sns.load_dataset("iris")

# 1. Scatter Plot: Sepal Length vs Petal Length
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
plt.title("Scatter Plot: Sepal vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()

# 2. Pairplot: Visual comparison of all features
sns.pairplot(iris, hue='species')
plt.suptitle("Iris Dataset - Pairplot", y=1.02)
plt.show()

# 3. Heatmap: Correlation between numerical features
correlation = iris.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap")
plt.show()
```
